```python
# 1. Percabangan / Branching

# Meminta input suhu dari pengguna
suhu = float(input("Masukkan suhu air pendingin (Â°C): "))

# Mengecek kondisi suhu
if suhu <= 90:
    print("STATUS: Normal")
elif suhu > 90 and suhu < 100:
    print("STATUS: Warning")
else:
    print("ALERT: SHUTDOWN")

```

```python
# 2. Fungsi / Functions

# Fungsi untuk menghitung efisiensi turbin
def hitung_efisiensi(input_energi, output_energi):
    efisiensi = output_energi / input_energi
    return efisiensi

# Contoh pemanggilan fungsi (Bonus)
hasil = hitung_efisiensi(500, 400)
print("Efisiensi Turbin:", hasil)
```

```python
#4. Logika if Majemuk/bertingkat

tekanan_uap = float(input("Masukkan tekanan uap (psi): "))
status_manual_override = False  # Misal override dalam keadaan OFF

# Logika if majemuk/bertingkat
if tekanan_uap > 150 and status_manual_override == False:
    print("Katup Terbuka")
else:
    print("Katup Tetap Tertutup")
```

Penjelasan Program 4

Program ini digunakan untuk menentukan apakah sebuah katup pengaman (safety valve) boleh terbuka berdasarkan kondisi tekanan uap dan status pengaturan manual. Dalam program ini, terdapat dua variabel utama, yaitu tekanan_uap yang bertipe float sebagai nilai tekanan uap dalam satuan psi, dan status_manual_override yang bertipe boolean untuk menunjukkan apakah katup sedang di-override secara manual atau tidak. Logika program menggunakan operator logika and, sehingga kedua syarat harus terpenuhi agar katup dapat terbuka. Kondisinya adalah sebagai berikut:

- Tekanan uap harus lebih besar dari 150 psi.

- status_manual_override harus bernilai False, artinya tidak ada override


Jika kedua kondisi tersebut benar, maka perintah print("Katup Terbuka") akan dijalankan, menandakan bahwa sistem mengizinkan katup terbuka secara otomatis. Namun jika salah satu atau kedua kondisi tidak terpenuhi, maka program akan mengeksekusi bagian else dan mencetak â€œKatup Tetap Tertutupâ€. Dengan logika ini, program membantu memastikan sistem keamanan bekerja dengan benar, hanya membuka katup ketika tekanan berbahaya dan tidak ada intervensi manual yang menahan katup tetap tertutup.

```python
# 5. Looping while

# Inisialisasi suhu awal
suhu_saat_ini = 60

# Loop akan berjalan selama suhu masih di bawah 85Â°C
while suhu_saat_ini < 85:
    print("Suhu saat ini:", suhu_saat_ini, "Â°C")
    # Naikkan suhu sebesar 5Â°C tiap iterasi
    suhu_saat_ini += 5

# Setelah loop selesai
print("Suhu telah mencapai atau melebihi 85Â°C.")
```

```python
# 7. Percabangan / Branching

# Program untuk mengklasifikasikan beban listrik

# Input data dari pengguna
kapasitas_maks = float(input("Masukkan kapasitas maksimum (Watt): "))
beban_terpakai = float(input("Masukkan beban terpakai (Watt): "))

# Hitung persentase pemakaian
persentase = (beban_terpakai / kapasitas_maks) * 100

# Klasifikasi berdasarkan persentase
if persentase <= 60:
    status = "Beban Ringan"
elif persentase <= 85:
    status = "Beban Sedang"
else:
    status = "Beban Berat"

# Tampilkan hasil
print(f"Persentase pemakaian: {persentase:.2f}%")
print("Status beban:", status)
```

Penjelasan Program 7

Program ini dibuat untuk mengklasifikasikan beban listrik berdasarkan persentase pemakaian terhadap kapasitas maksimum menggunakan struktur percabangan (if-elif-else).

Proses di dalam program dapat dijelaskan sebagai berikut:

- Penerimaan input pengguna: Program meminta pengguna memasukkan dua nilai, yaitu kapasitas_maks (kapasitas maksimum daya listrik) dan beban_terpakai (daya listrik yang sedang digunakan).

- Perhitungan persentase pemakaian: Persentase dihitung menggunakan rumus: persentase = (beban_terpakai / kapasitas_maks) * 100 Rumus ini menunjukkan seberapa besar beban terpakai dibandingkan dengan kapasitas maksimum sistem.

- Struktur percabangan: Program menggunakan if-elif-else untuk menentukan status beban: -Jika persentase â‰¤ 60, maka status â€œBeban Ringanâ€. -Jika persentase > 60 dan â‰¤ 85, maka status â€œBeban Sedangâ€. -Jika persentase > 85, maka status â€œBeban Beratâ€.

- Menampilkan hasil: Setelah klasifikasi selesai, program akan menampilkan hasil berupa persentase pemakaian dan status beban sesuai kondisi yang dihitung.

- Contoh hasil eksekusi: Jika pengguna memasukkan kapasitas_maks = 1000 dan beban_terpakai = 750, maka persentase pemakaian = 75%, dan program menampilkan status â€œBeban Sedangâ€.

Dengan program ini, pengguna dapat mengetahui tingkat penggunaan daya listrik secara efisien. Hal ini berguna untuk memantau kondisi sistem listrik agar tidak melebihi kapasitas maksimum dan tetap bekerja secara aman.

```python
# 8. Fungsi / Modularisasi

# Fungsi menghitung efisiensi termal turbin gas
def hitung_efisiensi_termal():
    # Input energi dari pengguna
    energi_masuk = float(input("Masukkan energi masuk (Joule): "))
    energi_keluar = float(input("Masukkan energi keluar (Joule): "))

    # Perhitungan efisiensi
    efisiensi = (energi_keluar / energi_masuk) * 100

    # Menampilkan hasil dengan format rapi
    print("\n===== HASIL PERHITUNGAN EFISIENSI =====")
    print(f"Energi Masuk       : {energi_masuk} Joule")
    print(f"Energi Keluar      : {energi_keluar} Joule")
    print(f"Efisiensi Termal   : {efisiensi:.2f}%")
    print("========================================")

# Memanggil fungsi
hitung_efisiensi_termal()

```

Penjelasan Program 8

Program ini digunakan untuk menghitung efisiensi termal pada turbin gas berdasarkan perbandingan antara energi masuk dan energi keluar. Di dalam fungsi hitung_efisiensi_termal(), pengguna akan diminta memasukkan dua nilai, yaitu energi_masuk dan energi_keluar, yang masing-masing bertipe float karena nilai energi dapat berupa bilangan desimal. Setelah nilai dimasukkan, program menghitung efisiensi menggunakan rumus Efisiensi (%) = (energi keluar / energi masuk) Ã— 100. Hasil perhitungan kemudian ditampilkan dengan format yang rapi, menampilkan energi masuk, energi keluar, serta nilai efisiensi yang sudah diformat dua angka di belakang koma.

Sebagai contoh, jika nilai energi masuk = 500 Joule dan energi keluar = 225 Joule, maka perhitungannya adalah:

- Efisiensi = (225 / 500) Ã— 100

- Efisiensi = 0,45 Ã— 100

- Efisiensi = 45%

Dengan demikian, program ini membantu dalam memahami efisiensi termal suatu turbin gas, yaitu seberapa besar energi yang berhasil dikonversi menjadi energi keluaran dibandingkan energi yang masuk, sehingga dapat digunakan sebagai parameter evaluasi performa sistem pembangkitan.

```python
# 9. Looping + List

suhu_boiler = [92.5, 88.7, 95.1, 102.3, 97.8]

for i in range(len(suhu_boiler)):
    print(f"Titik {i+1}: {suhu_boiler[i]}Â°C")
    if suhu_boiler[i] > 100:
        print(f"Peringatan: Suhu di titik {i+1} melebihiÂ 100Â°C")
```

Pada program ini saya membuat program untuk menampilkan hasil pembacaan suhu dari 5 titik di boiler. Di mana semua data suhunya itu disimpan di dalam list bernama suhu_boiler. Di dalam program, saya menggunakan perulangan for supaya programnya bisa menampilkan suhu dari setiap titik satu persatu.

Setelah suhu ditampilkan, program akan mengecek apakah ada suhu yang melebihi 100Â°C dengan menggunakan kondisi if. Kemudian jika ada suhu di titik tertentu lebih dari 100Â°C, maka programnya akan langsung menampilkan pesan peringatan bahwa suhu di titik boiler tersebut terlalu tinggi. Dari program ini bisa kita gunakan untuk menentukan di mana titik yang suhunya itu melewati batas aman, sehingga bisa dicek dan ditangani supaya boilernya tetap bekerja dengan baik dan aman.

```python
# 10. While Loop & Logika

# Simulasi pengisian tangki bahan bakar

tangki = 0  # mulai dari 0 liter

while tangki <= 100:
    print(f"Volume bahan bakar saat ini: {tangki} liter")

    # Kondisi hampir penuh
    if tangki == 80:
        print("WARNING: Hampir penuh!")

    # Kondisi penuh
    if tangki == 100:
        print("TANGKI PENUH â€“ Pengisian berhenti.")
        break  # menghentikan loop

    tangki += 10  # menambah 10 liter setiap iterasi
```

Penjelasan Program 10

Program ini digunakan untuk mensimulasikan proses pengisian tangki bahan bakar mulai dari kondisi kosong hingga penuh dengan kapasitas maksimum 100 liter. Variabel tangki diawali dengan nilai 0 liter sebagai kondisi awal. Selanjutnya, program menggunakan perulangan while yang berjalan selama nilai tangki kurang dari atau sama dengan 100. Pada setiap iterasi, program menampilkan volume bahan bakar saat ini kemudian menambahkan 10 liter menggunakan tangki += 10, sehingga level bahan bakar meningkat secara bertahap.

  
Di dalam loop terdapat dua kondisi pengecekan. Pertama, ketika volume mencapai 80 liter, program mencetak pesan "WARNING: Hampir penuh!" sebagai tanda bahwa kapasitas hampir maksimal dan proses pengisian perlu diperhatikan. Kedua, saat volume mencapai 100 liter, program menampilkan pesan "TANGKI PENUH â€“ Pengisian berhenti.", yang berarti tangki sudah mencapai kapasitas maksimum dan proses pengisian harus dihentikan. Pada kondisi ini, perintah break digunakan untuk menghentikan loop. Dengan simulasi sederhana ini, kita dapat memahami bagaimana logika perulangan dan pengecekan kondisi dapat diterapkan dalam sistem monitoring pengisian bahan bakar.

```python
# Challenge #

# ============================================
# Sistem Monitoring Energi PLTS Mini-Grid
# Menghitung produksi energi harian panel surya
# dan memantau status baterai
# ============================================

# Fungsi menghitung total energi panel surya
def hitung_total_energi(energi_per_jam):
    total = 0
    for energi in energi_per_jam:  # perulangan for
        total += energi
    return total

# Fungsi mengecek status baterai
def cek_status_baterai(kapasitas):
    if kapasitas > 80:
        return "Status: âœ… Baterai Penuh (Sistem Aman)"
    elif kapasitas > 40:
        return "Status: âš¡ Baterai Cukup (Normal)"
    elif kapasitas > 20:
        return "Status: ğŸŸ¡ Baterai Rendah (Perlu Perhatian)"
    else:
        return "Status: ğŸ”´ KRITIS! Segera Isi / Aktifkan Backup"

print("===== SISTEM MONITORING PLTS MINI-GRID =====")

# Input energi surya setiap jam (5 jam simulasi puncak matahari)
energi_jam = []
jam = 1

while jam <= 5:  # loop pengumpulan data
    energi = float(input(f"Masukkan energi panel jam ke-{jam} (kWh): "))
    energi_jam.append(energi)
    jam += 1

# Hitung total energi harian
total_energi = hitung_total_energi(energi_jam)
print(f"\nTotal Energi Surya Hari Ini: {total_energi:.2f} kWh")

# Input kapasitas baterai PLTS
kapasitas = float(input("Masukkan kapasitas baterai saat ini (%): "))
status_baterai = cek_status_baterai(kapasitas)

# Output Status
print("\n===== STATUS SISTEM =====")
print(f"Kapasitas Baterai: {kapasitas}%")
print(status_baterai)

# Logika operasional berdasarkan kondisi
if kapasitas <= 20:
    print("Rekomendasi: ğŸ”¥ Aktifkan genset cadangan / charging cepat")
elif total_energi < 15:
    print("Catatan: ğŸŒ¥ï¸ Produksi energi rendah, cuaca tidak optimal")
else:
    print("Sistem Operasional: âœ… Stabil & Optimal")
```
