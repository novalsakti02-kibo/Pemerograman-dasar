```python
#untuk memudahkan replacing maka di buatkan mapping jurusan sesuai dengan yg di inginkan, dengan kata-kata yg mirip di masukan ke jurusan yg seusai

mapping_jurusan = {
    "energi": "Pembangkit Energi",
    " energi ": "Pembangkit Energi",
    "ENERGI": "Pembangkit Energi",
    "EnErGi": "Pembangkit Energi",
    "eng": "Pembangkit Energi",
    "teknik": "Pembangkit Energi",
    "Teknik ": "Pembangkit Energi",
    "teknik ": "Pembangkit Energi",
    "manajemenenergi": "Manajemen Energi",
    "manajemen energi": "Manajemen Energi"
}
```

Penjelasan Langkah Langkah

- Membuat dictionary mapping_jurusan

  Kita membuat mapping berisi pasangan kata tidak konsisten → jurusan yang sudah distandarkan.

- Mengumpulkan variasi penulisan jurusan

  Semua bentuk seperti "energi", "ENERGI", "eng", "teknik", hingga "manajemen energi" dimasukkan sebagai kunci karena sering muncul dalam bentuk berbeda.

- Menentukan jurusan standar untuk tiap variasi

  Setiap variasi penulisan diarahkan ke bentuk rapi yang diinginkan, misalnya "Pembangkit Energi" atau "Manajemen Energi".

- Menyiapkan data untuk proses replace

  Mapping ini nantinya akan digunakan untuk mengganti seluruh teks jurusan menjadi bentuk yang konsisten hanya dengan satu perintah.

```python
SaktiKribo_df["jurusan"] = SaktiKribo_df["jurusan"].replace(mapping_jurusan)
```

Penjelasan Langkah-Langkah

- Mengubah seluruh isi kolom jurusan menjadi huruf kecil menggunakan .str.lower() agar penulisan konsisten dan tidak ada perbedaan karena kapitalisasi.

- Membersihkan spasi berlebih di awal atau akhir teks dengan .str.strip() supaya nilai yang sebenarnya sama tidak terbaca berbeda.

- Menyusun mapping berupa dictionary yang berisi variasi penulisan jurusan yang ingin diseragamkan. Mapping ini membantu mengelompokkan kata-kata yang mirip ke dalam satu kategori jurusan yang benar.

- Menerapkan mapping tersebut pada kolom jurusan menggunakan .replace(), sehingga seluruh variasi penulisan otomatis diganti menjadi format yang diinginkan.

- Melakukan pengecekan ulang menggunakan .value_counts() untuk memastikan bahwa seluruh penulisan jurusan sudah rapi, konsisten, dan tidak ada duplikasi penulisan.
