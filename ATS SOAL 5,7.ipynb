```python
suhu = float(input("Masukkan suhu air pendingin (°C): "))

if suhu <= 90:
    print("STATUS: Normal")
elif suhu < 100:
    print("STATUS: Warning")
else:
    print("ALERT: SHUTDOWN")
```

Penjelasan program nomor 1. Program ini membaca input suhu air pendingin dari pengguna dan menentukan status sistem menggunakan percabangan if-elif-else. Jika suhu ≤ 90°C, program menampilkan "STATUS: Normal". Jika suhu > 90°C dan < 100°C, ditampilkan "STATUS: Warning", dan jika suhu ≥ 100°C, program mencetak "ALERT: SHUTDOWN" sebagai tanda bahaya.

```python
vibrasi_mesin = [1.2, 0.9, 2.5, 3.1, 1.8]

for i in range(len(vibrasi_mesin)):
    if vibrasi_mesin[i] > 2.0:
        print(f"Warning: Unit {i+1} vibrasi tinggi!")
```

Penjelasan program nomor 3. Program ini memeriksa data vibrasi dari lima unit mesin menggunakan perulangan for. Jika nilai vibrasi suatu unit lebih dari 2.0, program menampilkan peringatan "Warning: Unit [nomor_unit] vibrasi tinggi!". Logika ini membantu mendeteksi unit mesin dengan getaran berlebih agar segera ditangani.
